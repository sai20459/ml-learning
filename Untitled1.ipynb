```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL
plt.plot(x,y)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("title")
# plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y,x,"b")
```

```python
#OOPS

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("x label")
axes.set_ylabel("y label")
axes.set_title("title")
```

```python
fig = plt.figure()
# [0.1,0.1,0.8,0.8] left bottom width height
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("larger plot")
axes2.plot(y,x)
axes2.set_title("smaller plot")
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2)

ax[0].plot(x,y)
ax[0].set_title("title")
ax[1].plot(y,x)

# ax.plot(x,y)
plt.tight_layout()
```

```python
# fg, ax = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

# ax[0].plot(x,y)
# ax[1].plot(y,x)


# plt.tight_layout()
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="x s")
ax.plot(x,x**3,label="x c")

ax.legend(loc=(0.1,0.1))
```

```python
fig.savefig("my_fig.png",dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# ax.plot(x,y,color="orange",linewidth=2,alpha=0.5,linestyle="-",marker="o",
#         markersize="10",markerfacecolor="pink",markeredgewidth=3,
#         markeredgecolor="green")
ax.plot(x,y,color="purple",linewidth=2,alpha=0.5,linestyle="--")
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
